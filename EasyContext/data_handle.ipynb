```python
import pandas as pd 
import json
data_path="/yzwl_data/yumu/clickmab_prompt_data.jsonl"
data=pd.read_json(data_path,lines=True)
# data.iloc[0]['prompt'][1]['content']
data
```

```python
def clean(x):
    for s in x:
        print(s)
        if isinstance(s['content'],list):
            s['content']=s['content'][0]["text"]
    return x 
data['prompt']=data['prompt'].map(clean)
```

```python
import json
df=pd.DataFrame()
def merge(row):
    return row['prompt']+[{"role":"assistant","content":row['response']}]
df["messages"]=data.apply(merge,axis=1)
df['answers']=data['response']+"<|im_end|>\n"
df.to_json("autoark_think.jsonl",lines=True,orient='records',force_ascii=False)
```
